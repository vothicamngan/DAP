```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
data = pd.read_csv("OnlineRetail.csv", encoding = "ISO-8859-1")
```

```python
data.info()
```

```python
data.describe()
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data = data[(data['UnitPrice']>=0) & (data['Quantity']>=0)]
```

```python
data
```

```python
data.describe()
```

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
```

```python
data['Year'] = data['InvoiceDate'].dt.year
```

```python
data['Month'] = data['InvoiceDate'].dt.month
```

```python
data['Amount'] = data['Quantity']* data['UnitPrice']
```

```python
data.head()
```

```python
data_2011 = data[(data['Year']==2011)]
```

```python
data_2011
```

```python
data_revenue = data_2011.groupby('Month')['Amount'].sum().reset_index()
data_revenue.columns = ['Month','Amount']
```

```python
data_2011 = data_2011[['InvoiceNo', 'InvoiceDate', 'Quantity', 'UnitPrice','Year','Month','Amount']]
```

```python
data_2011.drop_duplicates(subset='InvoiceNo',keep = 'first')
```

```python
data_total_order = data_2011.groupby('Month')['InvoiceNo'].count().reset_index()
data_total_order.columns = ['Month', 'InvoiceNo']
data_total_order
```

```python
x = data_total_order.index.get_level_values(0)
plt.bar(x,data_total_order['InvoiceNo'])
axes1 = plt.gca()
axes2 = axes1.twinx()
axes2.plot(x, data_revenue['Amount'], linewidth = 2, c = 'r', marker = 'o')
axes1.set_xlabel('Month', fontsize = 14)
axes1.set_ylabel('#Invoices', fontsize = 14)
axes2.set_ylabel('$', fontsize = 14)
axes1.legend(fontsize = 14)
axes2.legend(fontsize = 14)
plt.title('#Invoices and Revenue in 2011', fontsize = 16)
plt.show()
```
