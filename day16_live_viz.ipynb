# Trực quan hoá dữ liệu với Seaborn

```python
# Matplotlib
# plt.bar
# plt.plot
# plt.scatter
# plt.pie
```

```python
# 1 biến
# Phân bổ của dữ liệu - dữ liệu định lượng (numeric)
  # violinplot, histogram
# Đếm tần số của dữ liệu - dữ liệu định tính (categorical)
  # count plot

# 2 biến
# Mối quan hệ giữa 2 biến: scatterplot, lineplot
  # lmplot
# So sánh: range (phạm vi): box plot

# Nhiều hơn 2 biến:
# Mối quan giữa nhiều biến: heatmap

# Vẽ nhiều biểu đồ để thể hiện mối quan hệ giữa các biến: pairplot
```

## Set Up

```python
import matplotlib.pyplot as plt
import seaborn as sns #khai báo thư viện
%matplotlib inline
sns.set(style='whitegrid', palette='hsv')

import pandas as pd
import numpy as np
```

```python
# chuẩn bị dữ liệu
# nhà hàng
tips = sns.load_dataset("tips")
```

```python
# Xem nó
# 5 dòng đầu
tips.head()
```

```python
# tổng quan về nội dung dữ liệu
tips.describe()
```

## Biểu đồ thể hiện mối quan hệ giữa 2 biến định lượng: lmplot()

```python
# Vẽ biểu đồ mô tả mối quan hệ giữa total_bill và tip
# scatter plot
plt.scatter(x=tips['total_bill'], y=tips['tip'])
```

```python
# hàm lmplot của seaborn để thể hiện mối quan hệ giữa 2 biến định lượng: total_bill, và tip
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
# có sự khác biệt giữa người có hút thuốc và người không hút thuốc
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', col='smoker', data=tips)
```

```python
# Tổng kết
# Để biểu diễn mối quan hệ giữa 2 biến định lượng (numerical variables), có những cách sau
# cách 1: dùng matplotlib -> scatter plot
  # plt.scatter(x, y)
# cách 2: dùng lmplot của seaborn -> scatterplot + lineplot
  # sns.lmplot(x, y, data=)
  # Xem mối quan hệ giữa 2 biến định lượng theo nhóm trên cùng trục: sns.lmplot(x, y, hue=, data=)
  # Xem mối quan hệ giữa 2 biến định lượng theo nhóm - tách ra và đặt 2 hình cạnh nhau: sns.lmplot(x, y, col=, data=)
```

## Phân bổ của biến định lượng - Histogram, Violin, Box Plot

```python
# Histogram
# phân bổ của total_bill
tips['total_bill'].hist()

# tips['tip'].hist()
```

```python
# violinplot
sns.violinplot(y=tips['total_bill'])
```

```python
# So sánh phân bổ biến định lượng theo nhóm
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
# Xem phân bổ / range của biến định lượng
sns.boxplot(tips['total_bill'])
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# có sự khác nhau về phân bổ giữa nhóm hút thuốc và nhóm không hút thuốc không
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips)
```

```python
# có sự khác giữa giờ ăn hay không: time
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
```

```python
# Tổng kết
# Phân bổ của dữ liệu
# Cách 1: dùng histogram để xem phân bổ của 1 biến định lượng: df['total_bill'].hist()
# Cách 2: dùng violinplot

# Phân bổ của 1 biến định lượng theo nhóm:
# Ví dụ: xem phân bổ/khoảng của total_bill theo các ngày trong tuần
# Cách 1: dùng boxplot: sns.boxplot(x=, y=, data=)
# Cách 2: dùng violinplot: sns.violinplot(x=, y=, data=)
```

## Phân bổ của dữ liệu định tính

```python
# giới tính: Nam, Nữ
# đếm trong tập dữ liệu có bao nhiêu Nam, bao nhiêu Nữ
# countplot()
```

```python
sns.countplot(x='sex', data=tips)
```

```python
# Đếm bao nhiêu nam, bao nhiêu nữ, theo nhóm smoker
# bao nhiêu nam hút thuốc, bao nhiêu nam không hút thuốc
# bao nhiêu nữ hút thuốc, bao nhiêu nữ không hút thuốc
plt.figure(figsize=(10,10))
sns.countplot(y='sex', hue='smoker', data=tips)
plt.title('tên')
```

```python
# Tổng kết
# Xem phân bổ của dữ liệu định tính
# Đếm số lượng dữ liệu trong mỗi giá trị định tính
sns.countplot(...)
```

```python
# Vẽ phần chính -> seaborn
sns.scatterplot()
sns.lineplot()
sns.barplot()
sns.countplot()
sns.histplot()
sns.boxplot()
sns.heatmap()
sns.pairplot()
sns.lmplot()
# Vẽ trang trí -> matplotlib
# chỉ định kích cỡ của hình: plt.figure(figsize=())
# tên biểu đồ: plt.title('ten')
# tên trục x: plt.xlabel('ten truc x')
# tên trục y: plt.ylabel('ten truc y')
# Kết hợp biểu đồ:
  # Cùng trục
  # Cùng khung
```

## Heatmap

```python
import numpy as np
uniform_data = np.random.rand(10, 12)
sns.heatmap(uniform_data)
```

```python
corr = tips.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
df1 = tips[['tip', 'total_bill', 'size']].corr()
df1
sns.heatmap(df1, annot=True, cmap='coolwarm', fmt=".2f")
```

```python
# bước 1: .corr() -> mối tương quan giữa các biến -> df
# bước 2: từ bảng tương quan, vẽ heatmap
  # sns.heatmap(df)
```

```python
# Phân bổ của 1 biến
  # biến định lượng
  # biến định tính
# Mối quan hệ giữa 2 biến
# Mối quan hệ giữa nhiều biến -> tương quan
# Mỗi quan hệ giữa nhiều biến: phân bổ, mối quan hệ giữa các biến với nhau -> pairplot
```
