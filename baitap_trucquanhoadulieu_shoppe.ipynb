```python
data1 = data.groupby('join_year')['shopid'].count().reset_index()
data1.columns = ['join_year','count']
data1
```

```python
plt.bar(data1['join_year'],data1['count'])
plt.title('SỐ LƯỢNG SHOP GIA NHẬP THEO NĂM', color='r')
plt.xlabel("Năm")
plt.ylabel("Số lượng")
```

```python
plt.scatter(data['response_rate'],data['rating_good'])
plt.title('BIỂU ĐỒ THỂ HIỆN MỐI QUAN HỆ GIỮA TỶ LỆ ĐÁNH GIÁ VÀ ĐÁNH GIÁ TỐT', color='r')
plt.xlabel('Tỷ lệ đánh giá')
plt.ylabel('Đánh giá tốt')
```

```python
plt.scatter(data['response_time'], data['rating_bad'])
plt.title("BIỂU ĐỒ THỂ HIỂN MỖI QUAN HIỆN GIỮA THỜI GIAN VÀ MỨC ĐÁNH GIÁ XẤU (Đơn vị giây)", color="r")
plt.xlabel("Thời gian đánh giá")
plt.ylabel("Số lượt đánh giá xấu")
```
