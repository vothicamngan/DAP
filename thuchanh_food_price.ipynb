```python
data2 = data[(data['Year']==2019)&(data['Place']=='National Average')&(data['ProductName']=='Rice - Retail')]
```

```python
data2
```

```python
data_group = data2.groupby('Month')['Price'].mean().reset_index()
data_group.columns = ['Month','MeanPrice']
data_group
```

```python
plt.plot(data_group['Month'],data_group['MeanPrice'])
```
