```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime 
```

```python
data = pd.read_csv("OnlineRetail.csv", encoding = "ISO-8859-1")
```

```python
data.dropna(inplace=True)
```

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
```

```python
data['Month'] = data['InvoiceDate'].dt.month
```

```python
data
```

```python
data['Year']= data['InvoiceDate'].dt.year
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data1 = data[data['Year']==2011]
```

```python
data1
```

```python
data2 = data1[['Year','Month','InvoiceNo']]
```

```python
data2
```

```python
data2.drop_duplicates(subset = 'InvoiceNo', keep = 'first',inplace=True)
```

```python
data3 = data2.groupby('Month')['InvoiceNo'].count().reset_index()
data3.columns = ['Month','Count']
data3
```

```python
plt.plot(data3['Month'], data3['Count'])
```

```python
plt.bar(data3['Month'],data3['Count'])
```
