```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import RobustScaler
```

```python
y1 = np.random.normal(200, 30, size = 100)
y2 = np.random.normal(100,70, size = 100)
df = pd.DataFrame({'y1': y1, 'y2': y2, 'index': pd.date_range('1/1/2021', periods=100)})
```

```python
df.head()
```

```python
# V biu  ng biu din gi tr y1 theo thi gian (index)
plt.plot(df['index'], df['y1'], color='r')

# V biu  ng biu din gi tr y2 theo thi gian (index)
plt.plot(df['index'], df['y2'], color='b')

plt.show()
```

```python
# V biu  ng biu din gi tr y1 theo thi gian (index)
plt.plot(df['index'], df['y1'], color='r')
axes1 = plt.gca()
axes2 = axes1.twinx()
# V biu  ng biu din gi tr y2 theo thi gian (index)
axes2.plot(df['index'], df['y2'], color='b')
```
