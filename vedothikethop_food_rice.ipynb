```python
data_rice_2016 =data[(data['Year']==2016)&(data['Place']=='National Average')&(data['ProductName']=='Rice - Retail')]
data_rice_2017 =data[(data['Year']==2017)&(data['Place']=='National Average')&(data['ProductName']=='Rice - Retail')]
data_rice_2018 =data[(data['Year']==2018)&(data['Place']=='National Average')&(data['ProductName']=='Rice - Retail')]
data_rice_2019 =data[(data['Year']==2018)&(data['Place']=='National Average')&(data['ProductName']=='Rice - Retail')]
data_gas_2016 =data[(data['Year']==2016)&(data['Place']=='National Average')&(data['ProductName']=='Fuel (gas) - Retail')]
data_gas_2017 =data[(data['Year']==2017)&(data['Place']=='National Average')&(data['ProductName']=='Fuel (gas) - Retail')]
data_gas_2018 =data[(data['Year']==2018)&(data['Place']=='National Average')&(data['ProductName']=='Fuel (gas) - Retail')]
data_gas_2019 =data[(data['Year']==2018)&(data['Place']=='National Average')&(data['ProductName']=='Fuel (gas) - Retail')]
```
