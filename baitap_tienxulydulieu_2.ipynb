```python
import pandas as pd
import numpy as np
import seaborn as sns
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, RobustScaler, StandardScaler
```

```python
df = pd.read_csv("house_price_Dong-Da__subdata.csv")
```

```python
df.describe()
```

```python
df.isna()
```

```python
df = df.dropna(subset=['price'])
```

```python
df.describe()
```

```python
df
```

```python
df['land_certificate'].fillna('không có thông tin', inplace=True)
```

```python
df['house_direction'].fillna(df['house_direction'].mode().iloc[0], inplace=True)
```

```python
df['balcony_direction'].fillna(df['balcony_direction'].mode().iloc[0], inplace=True)
```

```python
df['toilet'].fillna(df['toilet'].mode().iloc[0], inplace=True)
```

```python
df['bedroom'].fillna(df['bedroom'].mode().iloc[0],inplace=True)
```

```python
df['floor'].fillna(df['floor'].mode().iloc[0], inplace=True)
```

```python
df
```

```python
df.describe()
```

```python
df_house_ngo = df[df['property_type']=='trong ngo']
df_house_ngo
```

```python
df['giá/m2'] = df['price']/df['area']
df
```

```python
df.describe()
```

```python
Q1_price = df['giá/m2'].quantile(0.25)
Q3_price = df['giá/m2'].quantile(0.75)
IQR_price = Q3_price- Q1_price
lower_price = Q1_price - 1.5*IQR_price
upper_price = Q3_price + 1.5*IQR_price
lower_price
```

```python
Q1_area = df['area'].quantile(0.25)
Q3_area = df['area'].quantile(0.75)
IQR_area = Q3_area-Q1_area
lower_area =Q1_area - 1.5*IQR_area
upper_area = Q3_area + 1.5*IQR_area
upper_area
```

```python
df = df[~((df['giá/m2']>upper_price)|(df['giá/m2']<lower_price))]
df
```

```python
df = df[~((df['area']>upper_area)|(df['area']<lower_area))]
```
