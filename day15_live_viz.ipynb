# Trực quan hoá dữ liệu FoodPrice in Turkey

```python
df = pd.read_csv("FoodPrice_in_Turkey.csv")
df.head()
```

```python
# Câu 1: Vẽ biểu đồ cột để so sánh giá Milk (powder, infant formula) và giá Fuel (gas)
# tại thời điểm tháng 12 năm 2019 ở Ankara, Istanbul, Izmir, và National Average
```

```python
df['Place'].unique()
```

```python
# Chuẩn bị dữ liệu
# 3 cột:
# Lọc dữ liệu tại tháng 12 năm 2019 ở Ankara, Istanbul, Izmir, và National Average
data = df[(df['Month'] == 12) & (df['Year'] == 2019)].reset_index(drop=True)

# Cột 2: Giá của sản phẩm Milk (powder, infant formula)
y1 = data[data['ProductName'] == 'Milk (powder, infant formula) - Retail'][['Place','Price']]
y1 = y1.reset_index(drop=True)
y1.columns = ['Place', 'Price_Milk']

# Cột 3: Giá Fuel (gas)
y2 = data[data['ProductName'] == 'Fuel (gas) - Retail'][['Place','Price']]
y2 = y2.reset_index(drop=True)
y2.columns = ['Place', 'Price_Fuel']

# Gộp thành 1 bảng
data1 = pd.merge(y1, y2, how='inner', on='Place')

# Vẽ
data1.plot(x = 'Place', y = ['Price_Milk', 'Price_Fuel'], kind = 'bar')
```

# Dùng dữ liệu OnlineRetail

```python
# Khai báo thư viện cần dùng:
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
# Load data
df = pd.read_csv("OnlineRetail.csv", encoding = "ISO-8859-1")
```

```python
# Biểu đồ cột thể hiện số lượng đơn hàng theo từng tháng của năm 2011. (hình 1)
# Biểu đồ đường thể hiện doanh thu theo từng tháng năm 2011. (hình 2)
# Kết hợp trục:
  # Trục 1: hình 1
  # Trục 2: hình 2
```

```python
#Biểu đồ cột
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['InvoiceYear'] = df['InvoiceDate'].dt.year
df['Revenue'] = df['UnitPrice']*df['Quantity']

# Lọc dữ liệu
df1 = df[df['InvoiceYear'] == 2011]
df2 = df1[['InvoiceNo', 'Revenue', 'InvoiceYear', 'InvoiceDate']].drop_duplicates(keep='first')
df2['InvoiceMonth'] = df2['InvoiceDate'].dt.month

# Gom nhóm dữ liệu và đếm số lượng đơn hàng theo tháng
df3 = df2.groupby('InvoiceMonth')['InvoiceNo'].count().reset_index()
df3.columns = ['Month', 'Count']
plt.bar(df3['Month'], df3['Count'])

# Chỉ định trục
axes1 = plt.gca()
axes2 = axes1.twinx()

# Vẽ biểu đồ đường thể hiện doanh thu theo tháng
df4 = df2.groupby('InvoiceMonth')['Revenue'].sum().reset_index()
df4.columns = ['Month', 'Revenue']
axes2.plot(df4['Month'], df4['Revenue'], color='r')
```
