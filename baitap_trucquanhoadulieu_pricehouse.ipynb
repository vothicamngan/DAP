```python
import pandas as pb
import numbers as num
import matplotlib.pyplot as plt
```

```python
data = pb.read_csv('house_price_Dong-Da__subdata.csv')
```

```python
data.info()
```

```python
data.isna()
```

```python
data_drop = data.dropna(subset='price')
```

```python
data_drop['land_certificate'].fillna('không có thông tin', inplace=True)
```

```python
data_drop['house_direction'].fillna(data_drop['house_direction'].mode().iloc[0], inplace=True)
```

```python
data_drop['balcony_direction'].fillna(data_drop['balcony_direction'].mode().iloc[0], inplace=True)
```

```python
data_drop['toilet'].fillna(data_drop['toilet'].mode().iloc[0], inplace=True)
```

```python
data_drop['bedroom'].fillna(data_drop['bedroom'].mode().iloc[0],inplace=True)
```

```python
data_drop['floor'].fillna(data_drop['floor'].mode().iloc[0], inplace=True)
```

```python
data_drop
```

```python
data_drop = data_drop[(data_drop['area']>0 )& (data_drop['price']>0)]
```

```python
data_drop = data_drop.dropna(subset='price')
```

```python
data_drop['giá/m2'] = data_drop['price']/data_drop['area']
```

```python
data_drop
```

```python
Q1_area = data_drop['area'].quantile(0.25)
Q3_area = data_drop['area'].quantile(0.75)
IQR_area = Q3_area-Q1_area
lower_area =Q1_area - 1.5*IQR_area
upper_area = Q3_area + 1.5*IQR_area
upper_area
```

```python
Q1_price = data_drop['price'].quantile(0.25)
Q3_price = data_drop['price'].quantile(0.75)
IQR_price = Q3_area-Q1_price
lower_price =Q1_price - 1.5*IQR_price
upper_price = Q3_price + 1.5*IQR_price
upper_price
```

```python
data_drop = data_drop[~((data_drop['area']>upper_area)|(data_drop['area']<lower_area))]
```

```python
data_drop
```
